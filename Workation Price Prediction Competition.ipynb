```python
#Encoding categorical variables
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
for col in cat_col:
    df[col] = lb.fit_transform(df[col])
```

```python
X = df.drop(columns=['Per Person Price','Travel Date'])
print('Shape:',X.shape)
X.head(2)
```

```python
y=df['Per Person Price']
```

```python
y.hist()
```

```python
from sklearn.preprocessing import RobustScaler
rs = RobustScaler()
for c in X.columns:
    X[c] = rs.fit_transform(X[c].values.reshape(-1, 1))
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20, random_state=30)
```

```python
# Training and testing with Regression models
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_log_error
mslr = mean_squared_log_error(y_pred,y_test)
np.log(mslr)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100,max_depth=1, min_samples_split=10)
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
mslr = mean_squared_log_error(y_pred,y_test)
np.log(mslr)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=200,max_depth=1, min_samples_split=9,bootstrap=False, oob_score=False, n_jobs=None, random_state=None,)
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
mslr = mean_squared_log_error(y_pred,y_test)
np.log(mslr)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(X_train,y_train)
y_pred = dt.predict(X_test)
mslr = mean_squared_log_error(y_pred,y_test)
np.log(mslr)
```
